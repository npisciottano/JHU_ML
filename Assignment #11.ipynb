```python
!pip3 install tensorflow keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD, RMSprop, Adagrad
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
n = 40
xx = np.random.random((n,1))
yy = np.random.random((n,1))

X = np.array([np.array([xx,-xx,-xx,xx]), np.array([yy,-yy,yy,-yy])]).reshape(2,4*n).T
Y = np.array([np.ones([2*n]), np.zeros([2*n])]).reshape(4*n)
X2 = X
Y2 = Y
X3 = X
Y3 = Y
X4 = X
Y4 = Y
X5 = X
Y5 = Y

plt.scatter(*zip(*X), c=Y)
```

```python
# One hidden layer
model = Sequential()

model.add(Dense(1, input_dim=2, activation='tanh'))
sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer='sgd')
model.fit(X, Y, batch_size=1, epochs=400)
print(model.predict_proba(X).reshape(4*n))

scores1 = model.evaluate(X,Y)
```

```python
# Two hidden layers
model = Sequential()

model.add(Dense(2, input_dim=2, activation='tanh'))
model.add(Dense(1, activation='tanh'))
sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer='sgd')
model.fit(X2, Y2, batch_size=1, epochs=400)
print(model.predict_proba(X).reshape(4*n))

scores2 = model.evaluate(X2,Y2)
```

```python
# Three hidden layers
model = Sequential()

model.add(Dense(2, input_dim=2, activation='tanh'))
model.add(Dense(4, activation='tanh'))
model.add(Dense(1, activation='tanh'))
sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer='sgd')
model.fit(X3, Y3, batch_size=1, epochs=400)
print(model.predict_proba(X).reshape(4*n))

scores3 = model.evaluate(X3,Y3)
```

```python
# Four hidden layers
model = Sequential()

model.add(Dense(2, input_dim=2, activation='tanh'))
model.add(Dense(8, activation='tanh'))
model.add(Dense(4, activation='tanh'))
model.add(Dense(1, activation='tanh'))
sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer='sgd')
model.fit(X4, Y4, batch_size=1, epochs=400)
print(model.predict_proba(X).reshape(4*n))

scores4 = model.evaluate(X4,Y4)
```

```python
# Five hidden layers
model = Sequential()

model.add(Dense(2, input_dim=2, activation='tanh'))
model.add(Dense(8, activation='tanh'))
model.add(Dense(16, activation='tanh'))
model.add(Dense(4, activation='tanh'))
model.add(Dense(1, activation='tanh'))
sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer='sgd')
model.fit(X5, Y5, batch_size=1, epochs=400)
print(model.predict_proba(X).reshape(4*n))

scores5 = model.evaluate(X5,Y5)
```

```python
plt.figure()
plt.xlabel('Number of Clusters')
plt.scatter(1,scores1)
plt.scatter(2, scores2)
plt.scatter(3, scores3)
plt.scatter(4, scores4)
plt.scatter(5, scores5)
```

```python
# Five hidden layers with sigmoid activators
model = Sequential()

model.add(Dense(2, input_dim=2, activation='sigmoid'))
model.add(Dense(8, activation='sigmoid'))
model.add(Dense(16, activation='sigmoid'))
model.add(Dense(4, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer='sgd')
model.fit(X5, Y5, batch_size=1, epochs=400)
print(model.predict_proba(X).reshape(4*n))

sigmoidscores5 = model.evaluate(X5,Y5)
```

```python
# Five hidden layers with different activators
model = Sequential()

model.add(Dense(2, input_dim=2, activation='exponential'))
model.add(Dense(8, activation='relu'))
model.add(Dense(16, activation='softplus'))
model.add(Dense(4, activation='elu'))
model.add(Dense(1, activation='softmax'))
sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer='sgd')
model.fit(X5, Y5, batch_size=1, epochs=400)
print(model.predict_proba(X).reshape(4*n))

differentscores5 = model.evaluate(X5,Y5)
```

```python
plt.figure()
plt.xlabel('Activators')
plt.scatter("tahn",scores5)
plt.scatter("sigmoid", sigmoidscores5)
plt.scatter("assorted", differentscores5)
```

```python
# Five hidden layers with different optimizer
model = Sequential()

model.add(Dense(2, input_dim=2, activation='exponential'))
model.add(Dense(8, activation='relu'))
model.add(Dense(16, activation='softplus'))
model.add(Dense(4, activation='elu'))
model.add(Dense(1, activation='softmax'))
rms = RMSprop(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer='RMSprop')
model.fit(X5, Y5, batch_size=1, epochs=400)
print(model.predict_proba(X).reshape(4*n))

rmsscores5 = model.evaluate(X5,Y5)
```

```python
# Five hidden layers with different optimizer
model = Sequential()

model.add(Dense(2, input_dim=2, activation='exponential'))
model.add(Dense(8, activation='relu'))
model.add(Dense(16, activation='softplus'))
model.add(Dense(4, activation='elu'))
model.add(Dense(1, activation='softmax'))
ada = Adagrad(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer='Adagrad')
model.fit(X5, Y5, batch_size=1, epochs=400)
print(model.predict_proba(X).reshape(4*n))

adascores5 = model.evaluate(X5,Y5)
```
