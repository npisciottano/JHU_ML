```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from sklearn import linear_model
```

```python
n = 100
x = np.linspace(0.01, 1, n).reshape(-1, 1)
y = np.linspace(0.01, 1, n)  + np.random.rand(n) - .5


plt.scatter(x,y)
```

## 1. Create and fit a Linear Regression Model
## Calculate the Training error and Testing error using sklearn with a .50 split 
For error, use `mean_squared`, but if you want to experiment with other mean errors, please do!

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.5)

model = LinearRegression()
model.fit(x_train, y_train)
model.coef_, model.intercept_

plt.scatter(x_train,y_train)
plt.plot(x_train, np.dot(x_train, model.coef_) + model.intercept_)
plt.scatter(x_test,y_test)

train=mean_squared_error(y_train, np.dot(x_train, model.coef_) + model.intercept_)
test=mean_squared_error(y_test, np.dot(x_test, model.coef_) + model.intercept_)

train,test
```

## 4. Chose an ideal split size based on the previous plot for Ridge.
## Vary the Ridge parameter `alpha` from 0 to any value you'd like above 1. Plot the Train and Test error. Describe what you see based on the `alpha` parameter's stiffness.

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

#Ideal split at bout 40% test, 60% train
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.40)

#An increase in alpha causes a stiffer line, reducing an overfit, but a very high level of alpha underfits the data
model = Ridge(alpha=15)
model.fit(x_train, y_train)
model.coef_, model.intercept_

train=mean_squared_error(y_train, np.dot(x_train, model.coef_) + model.intercept_)
test=mean_squared_error(y_test, np.dot(x_test, model.coef_) + model.intercept_)

plt.scatter(x_train,y_train)
plt.scatter(x_test, y_test)
plt.plot(x_train, np.dot(x_train, model.coef_) + model.intercept_)

train,test
```
