```python
import spacy
import pandas as pd
processor = spacy.load('en')
```

```python
processed_text = processor(text)
```

```python
#Tokenization
n = 0
df=[]
for sentence in processed_text.sents:
    for token in sentence:
        df.append({"n": n, "token":str(token), "POS":str(token.pos_), "Lemma":str(token.lemma_), "Is Stop":str(token.is_stop)})
        n += 1
```

```python
df2 = pd.DataFrame(df)
pd.DataFrame(df).head()
```

```python
#Group by all words
df2.groupby('token').count().sort_values(by=['n'],ascending=False).head()
```

```python
#Eliminate stop words, punctuation, and new lines
df3 = df2[(df2['Is Stop']=='False')]
df4 = df3[(df2['POS']!='PUNCT')]
df5 = df4[(df4['token']!="\n\n")]

#Group by non-stop words exclduing punctuation
df5.groupby('token').count().sort_values(by=['n'],ascending=False)
```

```python
#Group by non-stop words exclduing punctuation and part of speech
df5.groupby(['token','POS']).count().sort_values(by=['n'],ascending=False)
```

```python
#Dependency parsing

dpdf=[]
for sentence in processed_text.sents:
    for token in sentence:
        dpdf.append({"n": n, "token":str(token), "POS":str(token.pos_), "Lemma":str(token.lemma_), "Dependency":str(token.dep_),
                     "Is Stop":str(token.is_stop)})
        n += 1

dpdf2 = pd.DataFrame(dpdf)
pd.DataFrame(dpdf).head()
```

```python
def pr_tree(word,level):
    if word.is_punct:
        return
    for child in word.lefts:
        pr_tree(child, level+1)
    print('\t'* level + word.text + '-' + word.dep_)
    for child in word.rights:
        pr_tree(child, level+1)

for sentence in processed_text.sents:
    pr_tree(sentence.root, 0)
    print('***')
```

```python
for entity in processed_text.ents:
    print(entity, entity.label_,n)
```

```python
# Show the most common Entities and their types.
entdf=[]
for entity in processed_text.ents:
    entdf.append({"entity": entity, "label":entity.label_})
    n += 1

entdf2 = pd.DataFrame(entdf)
pd.DataFrame(entdf)
```

```python
entdf2.groupby('label').count().sort_values(by=['entity'],ascending=False)
```

```python
# Find Entites and their dependency
entdf2=[]
for entity in processed_text.ents:
    entdf2.append({"entity": entity, "label":entity.label_, "root":entity.root})
    n += 1

entdf3 = pd.DataFrame(entdf2)
pd.DataFrame(entdf3)
```

```python
for entity in processed_text.ents:
    pr_tree(entity.root.head, 0)
    print('******')
```
